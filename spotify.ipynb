```python
# Make second API call and collect album details
def get_album_details(album_id):
  url_album_details = "https://api.spotify.com/v1/albums/"+album_id+"?market=US"
  response_album_details = requests.get(url_album_details,
                          headers={"Authorization": f"Bearer {ACCESS_TOKEN}"}
                          ).json()

  album_popularity = response_album_details['popularity']
  album_total_tracks = response_album_details['total_tracks']
  album_label = response_album_details['label']

  return album_popularity,album_total_tracks,album_label
```

```python
def get_albums(df):
  # Make API call to collect Artist's Albums id, name, and release date
  url = "https://api.spotify.com/v1/artists/"+USHER_ID+"/albums?market=US"
  
  response = requests.get(url,
                          headers={"Authorization": f"Bearer {ACCESS_TOKEN}"},
                          ).json()

  for album in response["items"]:
    if album['album_type'] == 'album':
      album_id = album["id"]
      album_name = album["name"]
      release_date = album["release_date"]

      album_popularity,album_total_tracks,album_label = get_album_details(album_id)

      # Use pandas to store data
      df = df.append({'album_id':album_id, 
                    'album_name':album_name, 
                    'release_date':release_date,
                    'album_popularity':album_popularity,
                    'album_total_tracks':album_total_tracks,
                    'album_label':album_label}, ignore_index=True)

  return df
```

```python
#main

# build dataframe
df = pd.DataFrame(columns=['album_id','album_name','release_date','album_popularity','album_total_tracks','album_label'])

df = get_albums(df)
```
